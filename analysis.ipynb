```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
import jinja2
from IPython.display import display
from IPython.core.display import HTML
```

```python
df = pd.read_csv('game_list_2014.csv', index_col = 0, dtype={'match_id': object})
```

```python
matches = df.match_id.values
grounds = df.ground.values
dates = df.date.values
```

```python
df = pd.DataFrame()
for i in xrange(len(matches)):
    match = matches[i]
    ground = grounds[i]
    date = dates[i]
    df2 = pd.read_csv('ipl_2014_cleaned/%s.csv'%match)
    #df2['match_id'] = match
    #df2['ground_id'] = ground_id
    df2['ground'] = ground
    df2['date'] = date
    df = df.append(df2)
```

```python
df.drop(['ovr_ball_num', 'batsman_id', 'non_striker_id', 'bowler_id', 'player_out_id', 'game_end_time', 'match_id'], axis=1, inplace=True)
```

```python
df['batsman_dob'] = pd.to_datetime(df['batsman_dob'], format="%Y-%m-%d")
df['non_striker_dob'] = pd.to_datetime(df['non_striker_dob'], format="%Y-%m-%d")
df['bowler_dob'] = pd.to_datetime(df['bowler_dob'], format="%Y-%m-%d")
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
```

```python
def get_batter_age(s):
    if s['batsman_name'] == 'Rahul Shukla': return '1990-08-28'
    elif s['batsman_name'] == 'Aditya Tare': return '1987-11-07'
    elif s['batsman_name'] == 'Pragyan Ojha': return '1986-09-05'
    elif s['batsman_name'] == 'Yuzvendra Chahal': return '1990-07-23'
    elif s['batsman_name'] == 'Pawan Suyal': return '1989-11-10'
    else: return s['batsman_dob']
```

```python
def get_nonstriker_age(s):
    if s['non_striker_name'] == 'Rahul Shukla': return '1990-08-28'
    elif s['non_striker_name'] == 'Aditya Tare': return '1987-11-07'
    elif s['non_striker_name'] == 'Pragyan Ojha': return '1986-09-05'
    elif s['non_striker_name'] == 'Yuzvendra Chahal': return '1990-07-23'
    elif s['non_striker_name'] == 'Pawan Suyal': return '1989-11-10'
    else: return s['non_striker_dob']
```

```python
def get_bowler_age(s):
    if s['bowler_name'] == 'Rahul Shukla': return '1990-08-28'
    elif s['bowler_name'] == 'Aditya Tare': return '1987-11-07'
    elif s['bowler_name'] == 'Pragyan Ojha': return '1986-09-05'
    elif s['bowler_name'] == 'Yuzvendra Chahal': return '1990-07-23'
    elif s['bowler_name'] == 'Pawan Suyal': return '1989-11-10'
    else: return s['bowler_dob']
```

```python
df['batsman_dob'] = df.apply(get_batter_age, axis=1)
```

```python
df.columns
```

```python
df['non_striker_dob'] = df.apply(get_nonstriker_age, axis=1)
```

```python
df['bowler_dob'] = df.apply(get_bowler_age, axis=1)
```

```python
df['batsman_age'] = (df['date'] - df['batsman_dob']).dt.days/365
df['non_striker_age'] = (df['date'] - df['non_striker_dob']).dt.days/365
df['bowler_age'] = (df['date'] - df['bowler_dob']).dt.days/365
```

```python
df.drop(['batsman_dob', 'non_striker_dob', 'bowler_dob'], axis=1, inplace=True)
```

```python
pd.set_option('max_columns', 40)
```

```python
df['ball_speed'][df.ball_speed < 0] = np.nan
```

```python
df['landed_y'][df.landed_y > 50] = np.nan
```

```python
df['landed_x'][df.landed_x > 3] = np.nan
```

```python
df['ended_x'][df.ended_x > 3] = np.nan
```

```python
df.loc[(df.hit_x == 5) & (df.hit_y == 5), ['hit_x', 'hit_y']] = np.nan
```

```python
df.loc[(df.hit_x == 0) & (df.hit_y == 0), ['hit_x', 'hit_y']] = np.nan
```

```python
df.bat_right_handed = df.bat_right_handed == 'y'
```

```python
df['ended_x'][df.bat_right_handed == False] = -1*df['ended_x']
```

```python
df['landed_x'][df.bat_right_handed == False] = -1*df['landed_x']
```

```python
df.extras_type.unique()
```

```python
df['dummy_ball'] = (~((df.extras_type == 'Wd') | (df.extras_type == 'Nb') | (df.extras_type == 'NbLb'))).astype(int)
df['dot_ball'] = df['runs_w_extras'] == 0
df['single'] = df['runs_batter'] == 1
df['double'] = df['runs_batter'] == 2
df['triple'] = df['runs_batter'] == 3
df['four'] = (df['runs_batter'] == 4) | (df['runs_batter'] == 5)
df['six'] = df['runs_batter'] == 6
```

```python
df['dismissal_method'][df.dismissal_method == 'other'] = 'lbw'
```

```python
df['ball_speed'] = 3.60274248*df['ball_speed'] - 0.11928085
```

```python
def get_height(i):
    if i <=0.5: return 'full'
    elif i <=1.0: return 'good_length'
    else: return 'short'
```

```python
def get_line(i):
    if i <-0.6: return 'way_outside_off'
    elif i <-0.3: return 'outside_off'
    elif i <-0.15: return 'just_outside_off'
    elif i < 0: return 'on_off_stump'
    elif i <0.15: return 'on_leg_stump'
    else: return 'outside_leg'
```

```python
df['ball_height'] = df['ended_y'].apply(get_height)
df['ball_ended_line'] = df['ended_x'].apply(get_line)
```

```python
df['wicket'] = df['dismissal_method'].notnull().astype(int)
```

```python
df.dismissal_method.unique()
```

```python
df['bowler_wicket'] = (df['dismissal_method'].notnull()) & (df.dismissal_method != 'run out')
```

```python
df.columns
```

```python
batsmen_inning = df.groupby(('batsman_name', 'date')).agg('sum')[['dummy_ball', 'runs_batter', 'wicket', 'dot_ball', 'single', 'double', 'triple', 'four', 'six']].reset_index()
```

```python
bowler_inning = df.groupby(('bowler_name', 'date')).agg('sum')[['dummy_ball', 'runs_batter', 'runs_w_extras', 'bowler_wicket', 'dot_ball', 'single', 'double', 'triple', 'four', 'six']].reset_index()
```

```python
batsmen_agg = batsmen_inning.groupby('batsman_name').agg('sum')[['dummy_ball', 'runs_batter', 'wicket', 'dot_ball', 'single', 'double', 'triple', 'four', 'six']].join(batsmen_inning.groupby('batsman_name')[['date']].agg('count')).join(batsmen_inning.groupby('batsman_name')[['runs_batter']].agg('std'), rsuffix='_std')
```

```python
batsmen_agg.head()
```

```python
batsmen_agg.columns = ['total_balls', 'total_runs', 'num_out', 'dot_ball', 'single', 
                       'double', 'triple', 'four', 'six', 'tot_innings', 'std']
batsmen_agg['avg_inning'] = batsmen_agg['total_runs']/batsmen_agg['tot_innings']
batsmen_agg['runs_consistency'] = batsmen_agg['avg_inning']/batsmen_agg['std']
batsmen_agg = batsmen_agg[['tot_innings', 'total_balls', 'total_runs', 
                           'num_out', 'avg_inning', 'runs_consistency',
                           'dot_ball', 'single', 'double', 'triple', 'four', 'six']]
batsmen_agg['strike_rate'] = 100.*batsmen_agg['total_runs']/batsmen_agg['total_balls']
batsmen_agg['prop_dot'] = batsmen_agg['dot_ball']/batsmen_agg['total_balls']
batsmen_agg['prop_boundaries'] = (batsmen_agg['four'] + batsmen_agg['six'])/batsmen_agg['total_balls']
batsmen_agg['prop_running'] = (batsmen_agg['single'] + batsmen_agg['double']+ batsmen_agg['triple'])/batsmen_agg['total_balls']
batsmen_agg['runs_boundaries'] = (4*batsmen_agg['four'] + 6*batsmen_agg['six'])/batsmen_agg['total_runs']
batsmen_agg['runs_fours'] = (4*batsmen_agg['four'])/batsmen_agg['total_runs']
batsmen_agg['runs_sixes'] = (6*batsmen_agg['six'])/batsmen_agg['total_runs']
batsmen_agg['impact'] = batsmen_agg['strike_rate']*batsmen_agg['avg_inning']/100
```

```python
batsmen_agg[(batsmen_agg.tot_innings >= 5) & (batsmen_agg.total_balls > 50)].sort('prop_running', ascending=False).head(15)
```

```python
bowler_agg = bowler_inning.groupby('bowler_name').agg('sum')[['dummy_ball', 'runs_batter', 'runs_w_extras', 'bowler_wicket', 'dot_ball', 'single', 'double', 'triple', 'four', 'six']].join(df[['bowler_name', 'ball_speed']].groupby('bowler_name').agg('mean')).join(bowler_inning.groupby('bowler_name')[['date']].agg('count'))
```

```python
bowler_agg.columns = ['dummy_ball', 'runs_batter', 'runs_w_extras', 'bowler_wicket', 'dot_ball', 'single', 'double', 'triple', 'four', 'six', 'ball_speed', 'total_innings', u'strike_rate', u'average', u'economy_rate', u'prop_dot', u'prop_running', u'prop_boundaries', u'runs_boundaries', u'runs_running', u'runs_extras']
```

```python
bowler_agg['strike_rate'] = bowler_agg['dummy_ball']/bowler_agg['bowler_wicket']
bowler_agg['average'] = bowler_agg['runs_w_extras']/bowler_agg['bowler_wicket']
bowler_agg['economy_rate'] = 6.*bowler_agg['runs_w_extras']/bowler_agg['dummy_ball']
bowler_agg['prop_dot'] = bowler_agg['dot_ball']/bowler_agg['dummy_ball']
bowler_agg['prop_running'] = (bowler_agg['single'] + bowler_agg['double'] + bowler_agg['triple'])/bowler_agg['dummy_ball']
bowler_agg['prop_boundaries'] = (bowler_agg['four'] + bowler_agg['six'])/bowler_agg['dummy_ball']
bowler_agg['runs_boundaries'] = (4*bowler_agg['four'] + 6*bowler_agg['six'])/bowler_agg['runs_w_extras']
bowler_agg['runs_running'] = (1*bowler_agg['single'] + 2*bowler_agg['double'] + 3*bowler_agg['triple'])/bowler_agg['runs_w_extras']
bowler_agg['runs_extras'] = 1 - (bowler_agg['runs_batter'])/bowler_agg['runs_w_extras']
```

```python
bowler_agg = bowler_agg[['total_innings', 'dummy_ball', 'runs_batter', 'runs_w_extras', 'bowler_wicket',
                         'ball_speed', 'average', 'strike_rate', 'economy_rate', 'dot_ball', 'single',
                         'double', 'triple', 'four', 'six', 'prop_dot', 'prop_running', 'prop_boundaries',
                         'runs_boundaries', 'runs_running', 'runs_extras']]
```

```python
bowler_agg[bowler_agg.dummy_ball > 100].sort('bowler_wicket', ascending=False).head()
```

```python
ground_stats = df.groupby('ground').agg('sum')[['runs_batter', 'runs_w_extras', 'dummy_ball', 'bowler_wicket', 'dot_ball', 'single', 'double', 'triple', 'four', 'six']]
ground_stats['runs_boundaries'] = (4*ground_stats['four'] + 6*ground_stats['six'])/ground_stats['runs_w_extras']
ground_stats['dot_prop'] = (ground_stats['dot_ball'])/ground_stats['dummy_ball']
ground_stats['wicket_per_ball'] = (ground_stats['bowler_wicket'])/ground_stats['dummy_ball']
ground_stats['six_per_ball'] = (ground_stats['six'])/ground_stats['dummy_ball']
ground_stats['run_rate'] = 6.*ground_stats['runs_w_extras']/ground_stats['dummy_ball']
```

```python
from matplotlib.patches import Rectangle
```

```python
def plot_bowler_map(bowler):
    a = df[(df.bowler_name == bowler)]
    plt.figure(figsize=(3,5))
    plt.scatter(a['landed_x'], 22-a['landed_y'], alpha=0.05, s=150)
    plt.scatter(a[(a.bowler_wicket==True)]['landed_x'], 22-a[(a.bowler_wicket==True)]['landed_y'], alpha=0.2, color='red', s=150)
    plt.xlim(xmin=-1, xmax=1)
    plt.ylim(ymin=0, ymax=24)
    plt.title(bowler)
    plt.show()

    currentAxis = plt.gca()
    currentAxis.add_patch(Rectangle((-0.15, 0), 0.3, 0.731, facecolor="#999999", edgecolor='k', alpha=0.5))
    plt.scatter(a.ended_x, a.ended_y, alpha = 0.07, s=150)
    plt.scatter(a[(a.bowler_wicket == True)].ended_x, a[(a.bowler_wicket == True)].ended_y, alpha = 0.3, s=150, color='red')
    plt.ylim(ymin=0, ymax=2.5)
    plt.xlim(xmin=-1, xmax=1)
    plt.title(bowler)
    plt.show()

    plt.figure(figsize=(8,5))
    plt.gca().add_patch(Rectangle((-0.115, 0), 0.05, 0.731, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.gca().add_patch(Rectangle((-0.002, 0), 0.05, 0.731, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.gca().add_patch(Rectangle((0.115, 0), 0.05, 0.731, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.gca().add_patch(Rectangle((-0.115, 0.711), 0.28, 0.1, facecolor="#999999", edgecolor=None, alpha=0.7))
    plt.scatter(a[(a.bowler_wicket == True)].ended_x, a[(a.bowler_wicket == True)].ended_y, alpha = 0.4, s=150, color='red')
    plt.title('Bowling heatmap of %s, red dot represents wicket taking delivery'%bowler)
    a = a[['ended_x', 'ended_y']].dropna()    
    plt.hist2d(a['ended_x'].values, a['ended_y'].values, bins=(15,5), cmap='Oranges', 
               range=np.array([(-1, 1), (0, 2.5)]), alpha=0.5)
    plt.gca().grid(False)
    plt.show()
```
